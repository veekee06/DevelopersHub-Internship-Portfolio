Data visualization and exploratory data analysis help uncover relationships, distributions, and potential outliers within a dataset. This project aims to analyze the Iris dataset using summary statistics and visualizations.

## Dataset Understanding and Description
## Dataset Information
The Iris dataset contains measurements of iris flowers from three species:
i) Iris Setosa
ii) Iris Versicolor
iii) Iris Virginica

## Features
| Feature      | Description           |
| ------------ | --------------------- |
| Sepal Length | Length of sepal in cm |
| Sepal Width  | Width of sepal in cm  |
| Petal Length | Length of petal in cm |
| Petal Width  | Width of petal in cm  |
| Species      | Flower species        |


```python
# Import necessary libraries
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
# Load dataset
iris = pd.read_csv(r"C:\Users\UDDYBON\Downloads\iris.csv")
```

```python
# Display first five rows
iris.head()
```

## Data Cleaning and Preparation

```python
# Check dataset structure
iris.shape
iris.columns
iris.info()
```

```python
# Check missing values 
iris.isnull().sum()
```

```python
# Check duplicate records
iris.duplicated().sum()
```

## Observation 
No missing values found and no duplicate records found

## Exploratory Data Analysis

```python
# Generate descriptive statistics
iris.describe()
```

```python
# Relationship between sepal length and petal length
sns.scatterplot(data=iris,
                x='sepal_length',
                y='petal_length',
                hue='species')

plt.title("Sepal Length vs Petal Length")
plt.show()
```

```python
# Distribution of all numerical features
iris.hist(figsize=(10,8))
plt.show()
```

```python
# Detect outliers and spread of values
sns.boxplot(data=iris)
plt.show()
```

## Conclusion and Key Insights
## Conclusion
The Iris dataset was successfully explored and visualized using summary statistics and graphical techniques. The analysis revealed clear differences among the three flower species based on their sepal and petal measurements.

## Key Insights
1) Species Separation
* The scatter plot of Sepal Length vs Petal Length shows distinct clustering among the three species.
* Setosa is clearly separated from the other species, indicating that its petal characteristics are significantly different.
* Versicolor and Virginica show some overlap but remain reasonably distinguishable.

2) Feature Distribution
* Histograms indicate that most measurements follow approximately normal distributions.
* Petal-related measurements exhibit greater variation than some sepal measurements.

3) Outliers
* Boxplots show a few potential outliers in certain features.However, the overall dataset appears clean and well-structured.

## Overall Observation
Petal length and petal width appear to be the strongest features for distinguishing species.
